```
x=[4,7,8,9,11,5]
y=[13,23,5,7,9,7]
player=['virat','dhoni','vijay','dhawan']
runs=[120,44,77,33]
exp=[0.1,0,0,0.3]
```

```
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.bar(x,y)     #line plot only
plt.grid(color='green')
```

```
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.pie(runs,labels=player,explode=exp,shadow=True,autopct='%1.1f%%')
plt.grid(color='green')
```
